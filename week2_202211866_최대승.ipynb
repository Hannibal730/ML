* 기존의 출력데이터였던 1차원 벡터 x=[1,2,3]은 1차원 벡터Y=[5,6]으로 수정하였습니다.

```python
import numpy as np
import torch

# 데이터 읽기 함수
def load_dataset(file):
    data = np.loadtxt(file)
    print(f'수정한 데이터 DATA=\n{data}\n')

    # X에 입력 데이터, Y에 출력 데이터를 저장
    x = data[:,0:-1]  # 모든 행, 0열부터 마지막 열 직전 열
    print(f'입력데이터 x=\n{x}\n')

    y = np.reshape(data[:,-1],(data.shape[0], 1))  # 모든 행, 마지막 열을 가져오고, (2,1)로 재배열
    print(f'출력데이터 y=\n{y}\n')

    # 텐서로 변경
    x = torch.tensor(x, dtype=torch.float32)  # 입력 데이터를 텐서로 변환
    y = torch.tensor(y, dtype=torch.float32)  # 출력 데이터를 텐서로 변환

    return x, y

# 텐서를 리스트로 바꾸는 함수
def tensor2list(input_tensor):
    return input_tensor.cpu().detach().numpy().tolist()

# 데이터 불러오기
x, y = load_dataset("/gdrive/My Drive/ML/train2.txt")

# 가중치 초기화 (입력 데이터의 차원이 2이므로 가중치도 2개로 초기화)
w = torch.randn(x.size(1)) * 3.0  # x.size(1)은 입력 데이터의 열 개수
print("초기: w = {0}\n".format(tensor2list(w)))

# 학습률
alpha = 0.1

# 학습 데이터 수
m = y.size(0)

# 학습 루프
for epoch in range(50):
    cost = 0
    for i in range(m):
        # 가중치와 입력의 내적을 이용해 예측 값 계산
        cost = cost + ((torch.dot(w, x[i]) - y[i]) * x[i])
    cost = 1 / m * cost
    w = w - alpha * cost  # 가중치 업데이트

    # 학습 중 가중치 값 출력
    w1, w2 = tensor2list(w)[0], tensor2list(w)[1]
    print(f"Epoch {epoch + 1}: w1 = {w1}, w2 = {w2}")
```
